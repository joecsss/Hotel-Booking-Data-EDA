```python
import pandas as pd

df = pd.read_csv(r"C:\Users\User\Downloads\hotel_bookings.csv")

print(df.head())
print(df.info())
print(df.describe())
```

```python
#สำรวจ dataset เพื่อดู data structure และประเภทของข้อมูล เช่น booking date, country, lead time และ cancellation status
```

```python
cancellation_rate = df['is_canceled'].mean()

print("Cancellation Rate:", round(cancellation_rate*100,2), "%")
```

```python
#จากการวิเคราะห์พบว่า cancellation rate ของ booking อยู่ที่ประมาณ 37% ซึ่งอาจส่งผลต่อ revenue forecasting ของโรงแรม
```

```python
import matplotlib.pyplot as plt

bookings_by_month = df['arrival_date_month'].value_counts()

bookings_by_month.plot(kind='bar')
plt.title("Bookings by Month")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=45)
plt.show()
```

```python
#การวิเคราะห์ booking ตามเดือนช่วยให้เข้าใจ seasonal demand ของธุรกิจโรงแรม ซึ่งสามารถใช้ในการวาง pricing strategy หรือโปรโมชั่นในช่วง low season
```

```python
top_countries = df['country'].value_counts().head(10)

top_countries.plot(kind='bar')
plt.title("Top Customer Countries")
plt.xlabel("Country")
plt.ylabel("Bookings")
plt.show()
```

```python
#การวิเคราะห์ประเทศของลูกค้าช่วยให้ธุรกิจเข้าใจ market segment และสามารถปรับ marketing strategy ให้เหมาะกับกลุ่มลูกค้าหลัก
```

```python
df['lead_time'].hist(bins=30)

plt.title("Lead Time Distribution")
plt.xlabel("Days Before Arrival")
plt.ylabel("Number of Bookings")
plt.show()
```

```python
#lead time แสดงว่าลูกค้าจองล่วงหน้ากี่วันข้อมูลนี้สามารถช่วยในการวางแผน inventory ของห้องพักและ pricing strategy
```
